```python
import numpy as np
from sklearn.neighbors import KernelDensity
from scipy.signal import savgol_filter
_range = range


def gene_probabilities_histogram(gene_values, range=(-2, 2), bins=100, savgol_window=3):
    sizes, edges = np.histogram(gene_values, bins=bins, range=range, density=True)
    if savgol_window > 1:
        for _ in _range(3):
            sizes = savgol_filter(sizes, savgol_window, 1, mode="constant")
    return sizes, (edges[1:] + edges[:-1]) / 2


def gene_probabilities_kernel_density(
    gene_values,
    range=(-2, 2),
    bandwidth=2 / 100,  # width / bins = ~2*bandwidth
    kernel="gaussian",
    savgol_window=3,
):
    # fit density
    model = KernelDensity(bandwidth=bandwidth, kernel=kernel)
    model.fit(gene_values)

    # sample probabilities for a range of outcomes
    sample_values = np.arange(range[0], range[1] + bandwidth, bandwidth)
    probabilities = model.score_samples(sample_values[:, np.newaxis])
    probabilities = np.exp(probabilities)
    if savgol_window > 1:
        for _ in _range(3):
            probabilities = savgol_filter(probabilities, savgol_window, 1, mode="constant")
    return probabilities, sample_values


def cluster_gene_probabilities(gene_values, labels, normalized=False):
    probs, values = gene_probabilities_histogram(gene_values)
    cluster_probs = np.array([gene_probabilities_histogram(gene_values[labels == i])[0] for i in sorted(set(labels))])

    if normalized:
        # cluster_probs_normalized = np.divide(cluster_probs, probs, out=np.zeros_like(cluster_probs), where=probs != 0)
        cluster_probs_normalized = np.divide(
            cluster_probs,
            cluster_probs.sum(axis=0),
            out=np.zeros_like(cluster_probs),
            where=cluster_probs.sum(axis=0) != 0,
        )
        return cluster_probs_normalized, values

    return cluster_probs, values
```

```python
import matplotlib.pyplot as plt

gene0 = get_gene_values(X, 1000)

probs_hist, values_hist = gene_probabilities_histogram(gene0)
probs_kd, values_kd = gene_probabilities_kernel_density(gene0)

plt.plot(values_hist, probs_hist, color="blue")
plt.plot(values_kd, probs_kd, color="orange", alpha=0.6)
plt.show()
```

```python
from collections import defaultdict


def get_interesting_value_intervals(cluster_probs, values, confidence_value=0.8):

    # Get `cluster, values` indices for given `confidence_value`
    rows, cols = np.where(cluster_probs > confidence_value)

    # Group the interesting_values by cluster
    cluster_to_interesting_indices = defaultdict(list)
    for row, col in zip(rows, cols):
        cluster_to_interesting_indices[row].append(col)

    # Create intervals instead of single values
    cluster_to_intervals = defaultdict(list)
    for cluster, interesting_indices in cluster_to_interesting_indices.items():
        interesting_indices = sorted(interesting_indices)
        start = interesting_indices[0]

        for i in range(1, len(interesting_indices)):
            if interesting_indices[i - 1] + 1 != interesting_indices[i]:
                cluster_to_intervals[cluster].append((start, interesting_indices[i - 1]))
                start = interesting_indices[i]

        cluster_to_intervals[cluster].append((start, interesting_indices[-1]))

    # Lookup actual gene values from the indices
    cluster_to_interval_values = {
        cluster: list(map(lambda interval: (values[interval[0]], values[interval[1]]), list_of_intervals))
        for cluster, list_of_intervals in cluster_to_intervals.items()
    }
    return cluster_to_interval_values
```

```python
gene0 = get_gene_values(X, 0)
cluster_probs_normalized, values = cluster_gene_probabilities(gene0, l, normalized=True)
get_interesting_value_intervals(cluster_probs_normalized, values, confidence_value=0.8)
```

```python
from mpire.pool import WorkerPool
from multiprocessing import Manager


def add_interesting_value_intervals_to_dict(X, i, gene_cluster_intervals):
    gene_i = get_gene_values(X, i)
    cluster_probs_normalized, values = cluster_gene_probabilities(gene_i, l, normalized=True)
    interesting_value_intervals = get_interesting_value_intervals(cluster_probs_normalized, values, confidence_value=0.9)
    if interesting_value_intervals:
        gene_cluster_intervals[i] = interesting_value_intervals


with Manager() as manager:
    gene_cluster_intervals = manager.dict()
    pool = WorkerPool(use_dill=True)
    pool.map_unordered(
        lambda i: add_interesting_value_intervals_to_dict(X, i, gene_cluster_intervals),
        range(len(X[0])),
        progress_bar=True,
    )
    gene_cluster_intervals = gene_cluster_intervals.copy()
```
